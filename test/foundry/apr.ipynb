```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = []
b = []
c = []

year = 86400 * 365
step = 2000

with open('kek.logs', 'r') as f:
    z = f.readlines()
    for i in range(len(z)):
        v = str(z[i]).split(' ')
        a.append(int(v[0]) / 10**6)
        b.append(int(v[1].split('\n')[0]))
        
        c.append(100 * (int(v[0]) - 10**8) * year / ((i + 1) * step) / 10**8)
        
plt.plot(a)

plt.show()

plt.plot(b)
plt.show()

plt.plot(c)
plt.show()
        
```
